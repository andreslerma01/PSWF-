```python
import numpy as np
from scipy.special import pro_ang1
import matplotlib.pyplot as plt

# Define parameters
n = 0  # Order of the PSAF
k = 16  # Azimuthal angle (set to a constant for visualization)
theta = np.linspace(-1, 1, 500)  # Theta values
theta = theta[1:-1]
delta_theta = theta[1] - theta[0]  # Calculate spacing (Delta x)


# Values of alpha to visualize
alpha_values = [0, 2, 4, 6]

# Create a plot
plt.figure(figsize=(12, 6))

# Loop through each alpha value and plot the corresponding PSAF
for alpha in alpha_values:
    psaf_values = pro_ang1(n, alpha, phi, theta)[0]  # Compute PSAF values
    
    # Calculate the L2 norm
    l2_norm = np.sqrt(np.sum(np.abs(psaf_values) ** 2) * delta_theta)/2

    # Normalize the function using its L2 norm
    psaf_values_normalized = psaf_values / l2_norm

    
    plt.plot(theta, psaf_values_normalized, label=fr'$\varphi_{{16,{alpha}}}$')

    

plt.legend(
    fontsize=26,  # Increase legend font size
    loc='upper center',  # Place legend at the top center of the figure
    bbox_to_anchor=(0.5, -0.1),  # Adjust position below the plot
    ncol=4  # Arrange the labels in a single row (or adjust as needed)
)

plt.grid(True)
plt.xlim(-1, 1)  # Restrict x-axis to the range -1 to 1

plt.ylim(-2.1, 3.1)  # Restrict y-axis to the range -1 to 1



# Increase font size of axis tick labels
plt.tick_params(axis='both', which='major', labelsize=26)  # Major ticks
plt.tick_params(axis='both', which='minor', labelsize=26)  # Minor ticks


# Save the plot as a PDF
plt.savefig("psaf_plot.pdf", format="pdf", bbox_inches="tight")

# Show the plot
plt.show()
```
